```python
l=[]
def rolling(x,l):
    for i in range(2,24):
        x['co_rolling'+str(i)]=x['CO'].rolling(window=i,min_periods=1).mean()
        y=x[x['CO'].notnull()]
        l.append(mean_squared_error(y['CO'],y['co_rolling'+str(i)], squared = False))
    print(min(l),"index :",l.index(min(l))+2)
```
